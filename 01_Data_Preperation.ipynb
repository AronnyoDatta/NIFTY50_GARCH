# Data Preparation â€“ NIFTY 50

This notebook prepares NIFTY 50 historical price data for time-series and
volatility modeling. The output of this step is a clean, chronologically
ordered price series used throughout the project.

```python
import pandas as pd
```

## Load Data

Import historical NIFTY 50 price data from Excel.

```python
df=pd.read_excel("Nifty 50 Historical Data (1).xlsx")
```

```python
df.head()
```

## Date Formatting

Convert the `Date` column to datetime format for time-series operations.

```python
df['Date'] = pd.to_datetime(df['Date'])
```

## Variable Selection

Retain only date and closing price, which are required for return and volatility analysis.

```python
df = df[['Date', 'Price']].copy()
df.head()
```

## Sorting

Sort the data chronologically to ensure correct temporal ordering.

```python
df = df.sort_values(by='Date').reset_index(drop=True)
df.head()

```
