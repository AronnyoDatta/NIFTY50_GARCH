```python
import pandas as pd

df = pd.read_excel("Nifty 50 Historical Data (1).xlsx")
df = df[['Date', 'Price']].copy()
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date').reset_index(drop=True)
```

## Price Trend Visualization

A line plot is used to visualize the long-term movement of the NIFTY 50
index over the sample period.

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Price')

plt.xlabel('Date')
plt.ylabel('NIFTY 50 Price')
plt.title('NIFTY 50 Price Trend')

plt.show()
```

## Interpretation

The plot provides a visual overview of trend behavior, structural shifts,
and major market movements, serving as contextual groundwork for
return and volatility modeling.
